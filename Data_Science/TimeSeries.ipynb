```python
import pandas as pd
import plotly.io as pio
from IPython.display import display, HTML
pio.renderers.default = "notebook_connected"
pd.options.plotting.backend = "plotly"
style = f"""
<style>
.container {{ width:100% !important; }}
div.output_area {{
    overflow-x: scroll;
}}
div.output_area img {{
    max-width: 500;
}}
</style>
"""
display(HTML(style))
```

```python
df = pd.read_csv("resources/datasets/UVA001_DISC_tseries.csv")["Disc"]
df = (df - df.min()) / (df.max() - df.min())

means = [10, 20, 40, 80, 150]

import plotly.graph_objects as pgo
fig = pgo.Figure()

for i,m in enumerate(means):
    _offset = (i / len(means))*2
    data = df + _offset
    ravg = data.rolling(window=m, center=True).mean()

    fig.add_trace(pgo.Scatter(x=df.index, y=data, mode='lines', marker_color="grey", name=None))
    fig.add_trace(pgo.Scatter(x=df.index, y=ravg, mode='lines', name=f'Window {m}'))


fig.update_layout(
    xaxis={"rangeslider": {"visible": True}},
    # autosize=False,
    # height=800,
    template="plotly_dark",
    # width=5000
)


fig.show()
```
